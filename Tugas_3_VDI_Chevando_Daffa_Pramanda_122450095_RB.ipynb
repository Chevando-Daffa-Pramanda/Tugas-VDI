**Data yang digunakan**

Disini saya menggunakan data Harga Produsen Ketela Pohon pada tahun 2013 dan 2014 berdasarkan Badan Pusat Statistik Provinsi Lampung

**Tujuan**

Tujuan utama dari visualisasi ini adalah memberikan gambaran perbandingan harga antara tahun 2013 dan tahun 2014 selama periode 12 bulan. Hal ini dapat digunakan untuk Melihat bagaimana harga naik atau turun secara bertahap atau tiba-tiba dari bulan ke bulan dan dapat juga menentukan apakah ada pola atau musim tertentu yang mempengaruhi pergerakan harga.

**User**

* Pengambil keputusan: Pengelola bisnis atau pemangku kepentingan yang membutuhkan informasi untuk membuat keputusan terkait harga, produksi, atau distribusi berdasarkan data historis.
* Investor atau pelaku pasar: Orang yang tertarik untuk memahami perilaku harga untuk tujuan investasi atau pembelian.

**Fungsi**

* Perbandingan: Memungkinkan pengguna untuk membandingkan dengan jelas antara Harga tahun 2013 dan Harga tahun 2014 berdasarkan per bulannya.
* Analisis tren: Membantu dalam mengidentifikasi tren atau fluktuasi musiman yang mungkin tidak terlihat jelas dalam tabel data.
* Mendeteksi anomali: Visualisasi juga dapat digunakan untuk mendeteksi anomali atau penyimpangan yang signifikan dalam harga.

**Tone**

* Informatif dan profesional: Visualisasi ini dirancang untuk menyampaikan informasi secara jelas, ringkas, dan objektif. Dengan demikian, tone yang digunakan adalah profesional dan berbasis data.
* Netral:  Data harga disajikan tanpa mengindikasikan penilaian positif atau negatif terhadap fluktuasi harga, melainkan hanya menggambarkan fakta dari perkembangan harga tersebut.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Bulan' : ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Harga 2013' : [77000,75000,83000,84000,90000,88000,90000,95000,87500,80000,75000,75000],
    'Harga 2014' : [80000,80000,75000,72500,72500,72500,72500,98000,98000,100000,85000,72000]

}

# data frame
df = pd.DataFrame(data)

# memvisualisasikan menggunakan bar chart
plt.figure(figsize=(10, 6))

bar_width = 0.35
index = range(len(df['Bulan']))

plt.bar(index, df['Harga 2013'], width=bar_width, label='Harga 2013', color='brown', align='center')
plt.bar([i + bar_width for i in index], df['Harga 2014'], width=bar_width, label='Harga 2014', color='grey', align='center')

plt.title('Perbandingan Produsen Ketela Pohon Harga tahun 2013 dan Harga tahun 2014 per Bulan', fontsize=12)
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Harga dalam rupiah', fontsize=12)

plt.xticks([i + bar_width / 2 for i in index], df['Bulan'], rotation=25)

plt.grid(True, axis='y')
plt.legend()

plt.tight_layout()
plt.show()
```
